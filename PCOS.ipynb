# Preprocessing

```python
import pandas as pd
```

```python
path = '/content/drive/MyDrive/Data Science Lab Data/pcos_data.xlsx'
```

```python
data = pd.read_excel(path)
```

```python
data.head()
```

```python
data.drop(data.columns[44], axis=1, inplace=True)
data.drop('Sl. No', axis=1, inplace=True)
data.drop('Patient File No.', axis=1, inplace=True)
```

```python
data.drop('Marraige Status (Yrs)', axis=1, inplace=True)
```

```python
data.isnull().sum()
```

```python
data.columns = ['PCOS (Y/N)', 'Age (yrs)', 'Weight (kg)', 'Height (cm)', 'BMI', 'Blood Group', 'Pulse rate (bpm)', 'RR (breaths/min)', 'Hb (g/dl)', 'Cycle (R/I)', 'Cycle length (days)', 'Pregnant (Y/N)', 'No. of abortions', 'I beta-HCG (mIU/mL)', 'II beta-HCG (mIU/mL)', 'FSH (mIU/mL)', 'LH (mIU/mL)', 'FSH/LH ratio', 'Hip (inch)', 'Waist (inch)', 'Waist/hip ratio', 'TSH (mIU/L)', 'AMH (ng/mL)', 'PRL (ng/mL)', 'Vitamin D3 (ng/mL)', 'PRG (ng/mL)', 'RBS (mg/dl)', 'Weight gain (Y/N)', 'Hair growth (Y/N)', 'Skin darkening (Y/N)', 'Hair loss (Y/N)', 'Pimples (Y/N)', 'Fast food (Y/N)', 'Regular exercise (Y/N)', 'BP systolic (mmHg)', 'BP diastolic (mmHg)', 'Follicle No. (L)', 'Follicle No. (R)', 'Avg. follicle size (L) (mm)', 'Avg. follicle size (R) (mm)', 'Endometrium (mm)']
```

## Calculating missing values

```python
for i in range(541):
  height_meters = data['Height (cm)'][i]/100
  data['BMI'][i] = round((data['Weight (kg)'][i] / (height_meters*height_meters)), 2)
```

```python
for i in range(541):
  data['FSH/LH ratio'][i] = data['FSH (mIU/mL)'][i]/data['LH (mIU/mL)'][i]
```

```python
for i in range(541):
  data['Waist/hip ratio'][i] = data['Waist (inch)'][i]/data['Hip (inch)'][i]
```

## Imputations

```python
data['Fast food (Y/N)'][156] = 0
```

```python
import statistics
```

```python
data['AMH (ng/mL)'][305] = statistics.mean(data['AMH (ng/mL)'][:304])
```

```python
data.isnull().sum()
```

## Casting values

```python
data['II beta-HCG (mIU/mL)'][123] = float(1.99)
```

```python
data['II beta-HCG (mIU/mL)'] = data['II beta-HCG (mIU/mL)'].astype('float')
data['AMH (ng/mL)'] = data['AMH (ng/mL)'].astype('float')
```

## Neural networks

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.losses import MeanSquaredError
from keras.optimizers import Adam
```

```python
model1 = Sequential()

model1.add(Dense(16, input_dim=16, activation='relu'))
model1.add(Dropout(0.1, input_shape=(16,)))

model1.add(Dense(8, activation='relu'))
model1.add(Dropout(0.1, input_shape=(8,)))

model1.add(Dense(4, activation='relu'))
model1.add(Dropout(0.1, input_shape=(8,)))

model1.add(Dense(4, activation='relu'))

model1.add(Dense(1, activation='sigmoid'))

model1.summary()
```

```python
model1.compile(optimizer='adam', loss='binary_crossentropy', metrics = ['accuracy'])
```

```python
history = model1.fit(X_train, y_train, epochs=150, batch_size=10, validation_split=0.2, shuffle=True, verbose=1)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```
